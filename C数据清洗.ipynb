```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体或其他您系统中存在的中文字体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号'-'显示为方块的问题

# 读取数据
df = pd.read_csv('附件.csv', encoding='utf-8')

# 查看数据基本信息
print("数据形状:", df.shape)
print("\n前5行数据:")
display(df.head())

print("\n数据基本信息:")
df.info()

print("\n数值列描述性统计:")
display(df.describe())

# 检查缺失值
print("\n每列缺失值数量:")
print(df.isnull().sum())

# 数据清洗步骤
# 1. 提取孕周数值（将"11w+6"这样的格式转换为周数）
def extract_weeks(gestational_week):
    if pd.isna(gestational_week):
        return np.nan
    if isinstance(gestational_week, str):
        # 匹配"数字w+数字"或"数字w"的格式
        match = re.search(r'(\d+)w(?:\+(\d+))?', gestational_week)
        if match:
            weeks = int(match.group(1))
            days = int(match.group(2)) if match.group(2) else 0
            return weeks + days/7
    return np.nan

# 应用转换
df['检测孕周_数值'] = df['检测孕周'].apply(extract_weeks)

# 查看转换结果
print("\n孕周转换结果示例:")
print(df[['检测孕周', '检测孕周_数值']].head(10))

print("\n检测孕周数值的描述性统计:")
print(df['检测孕周_数值'].describe())

# 2. 清洗条件：Y染色体浓度 ≤ 4% 且 孕期在10-25周之间
print(f"\n清洗前样本数: {len(df)}")

# 条件1: Y染色体浓度 ≤ 4%
condition1 = df['Y染色体浓度'] >= 0.04

# 条件2: 孕期在10-25周之间
condition2 = (df['检测孕周_数值'] >= 10) & (df['检测孕周_数值'] <= 25)

# 应用两个条件
df_cleaned = df[condition1 & condition2].copy()

print(f"剔除Y染色体浓度>4%后样本数: {len(df[condition1])}")
print(f"孕期在10-25周之间的样本数: {len(df[condition2])}")
print(f"同时满足两个条件的样本数: {len(df_cleaned)}")

# 查看被剔除的样本原因
df['剔除原因'] = '保留'
df.loc[~condition1, '剔除原因'] = 'Y浓度<4%'
df.loc[~condition2, '剔除原因'] = '孕周不在10-25周'
df.loc[~(condition1 | condition2), '剔除原因'] = '两个条件都不满足'

print("\n样本剔除情况统计:")
print(df['剔除原因'].value_counts())


# 2. 处理其他可能的异常值（可选：例如Z值绝对值大于5的极端值）
z_columns = ['13号染色体的Z值', '18号染色体的Z值', '21号染色体的Z值', 'X染色体的Z值', 'Y染色体的Z值']
for col in z_columns:
    extreme_count = (df_cleaned[col].abs() > 5).sum()
    print(f"{col} 绝对值大于5的样本数: {extreme_count}")

# 3. 处理重复值（如果有）
duplicate_count = df_cleaned.duplicated().sum()
print(f"重复样本数: {duplicate_count}")
if duplicate_count > 0:
    df_cleaned = df_cleaned.drop_duplicates()

# 4. 处理缺失值（根据实际情况选择填充或删除）
# 这里仅做展示，实际可根据业务需求处理
print("清洗后缺失值情况:")
print(df_cleaned.isnull().sum())

# 可视化部分

# 1. Y染色体浓度分布图（清洗前后对比）
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 清洗前
axes[0].hist(df['Y染色体浓度'].dropna(), bins=30, alpha=0.7, color='red', edgecolor='black')
axes[0].axvline(0.04, color='blue', linestyle='--', label='阈值 (6%)')
axes[0].set_title('清洗前 Y染色体浓度分布')
axes[0].set_xlabel('Y染色体浓度')
axes[0].set_ylabel('频数')
axes[0].legend()

# 清洗后
axes[1].hist(df_cleaned['Y染色体浓度'].dropna(), bins=30, alpha=0.7, color='green', edgecolor='black')
axes[1].axvline(0.04, color='blue', linestyle='--', label='阈值 (6%)')
axes[1].set_title('清洗后 Y染色体浓度分布')
axes[1].set_xlabel('Y染色体浓度')
axes[1].set_ylabel('频数')
axes[1].legend()

plt.tight_layout()
plt.show()

# 2. 各染色体Z值分布箱线图（清洗后）
z_columns = ['13号染色体的Z值', '18号染色体的Z值', '21号染色体的Z值', 'X染色体的Z值', 'Y染色体的Z值']
plt.figure(figsize=(10, 6))
df_cleaned[z_columns].boxplot()
plt.title('清洗后各染色体Z值分布')
plt.xticks(rotation=45)
plt.ylabel('Z值')
plt.show()

# 3. 胎儿健康情况分布
if '胎儿是否健康' in df_cleaned.columns:
    health_count = df_cleaned['胎儿是否健康'].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(health_count, labels=health_count.index, autopct='%1.1f%%', startangle=90)
    plt.title('胎儿健康情况分布（清洗后）')
    plt.show()

# 4. 孕妇BMI分布
if '孕妇BMI' in df_cleaned.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df_cleaned['孕妇BMI'].dropna(), kde=True, bins=30)
    plt.title('孕妇BMI分布（清洗后）')
    plt.xlabel('BMI')
    plt.ylabel('频数')
    plt.show()

# 5. 年龄分布
if '年龄' in df_cleaned.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df_cleaned['年龄'].dropna(), kde=True, bins=20)
    plt.title('孕妇年龄分布（清洗后）')
    plt.xlabel('年龄')
    plt.ylabel('频数')
    plt.show()

# 6. 相关性热图（数值型变量）
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12, 10))
sns.heatmap(df_cleaned[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('数值变量相关性热图（清洗后）')
plt.show()

# 保存清洗后的数据（可选）
df_cleaned.to_csv('cleaned_data.csv', index=False, encoding='utf-8-sig')
print("清洗后的数据已保存为 cleaned_data.csv")
```

# 修正版

```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re  # 修复：之前未导入 re

# 设置中文字体和绘图样式
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体或其他您系统中存在的中文字体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号'-'显示为方块的问题

# ==================== 1. 读取数据 ====================
df = pd.read_csv('附件.csv', encoding='utf-8')

print("原始数据形状:", df.shape)
print("\n前5行数据:")
display(df.head())

print("\n数据基本信息:")
df.info()

print("\n数值列描述性统计:")
display(df.describe())

print("\n每列缺失值数量:")
print(df.isnull().sum())
```

```python
# 假设存在“检测日期”列，转换为标准 datetime 格式
if '检测日期' in df.columns:
    df['检测日期'] = pd.to_datetime(df['检测日期'], errors='coerce')  # 转换失败设为 NaT
    print(f"检测日期中无效日期数量: {df['检测日期'].isnull().sum()}")
```

```python
def extract_weeks(gestational_week):
    """提取孕周：支持 '11w+6', '11w', '11周6天', '11周' 等格式"""
    if pd.isna(gestational_week):
        return np.nan
    if isinstance(gestational_week, (int, float)):
        return float(gestational_week)  # 已是数值型，直接返回

    # 统一转为字符串
    gestational_week = str(gestational_week).strip()

    # 匹配模式：11w+6, 11w, 11周6天, 11周
    patterns = [
        r'(\d+)w\+(\d+)',           # 11w+6
        r'(\d+)w',                  # 11w
        r'(\d+)周(?:\s*(\d+)天)?'   # 11周 或 11周6天
    ]

    for pattern in patterns:
        match = re.search(pattern, gestational_week)
        if match:
            weeks = int(match.group(1))
            days = int(match.group(2)) if match.lastindex == 2 else 0
            return round(weeks + days / 7, 2)

    # 若无法匹配，尝试提取第一个数字作为周数
    num_match = re.search(r'\d+', gestational_week)
    if num_match:
        return float(num_match.group())

    return np.nan

# 应用转换
df['检测孕周_数值'] = df['检测孕周'].apply(extract_weeks)
print("\n孕周转换结果示例:")
print(df[['检测孕周', '检测孕周_数值']].head(10))

print("\n检测孕周数值的描述性统计:")
print(df['检测孕周_数值'].describe())
```

```python
print(f"\n清洗前样本数: {len(df)}")

# 清洗条件1: Y染色体浓度 ≥ 4% （注意：原文是“≤ 4%”，但通常NIPT中Y>0表示男胎，>4%才可信）
# 根据业务逻辑修正：此处应为 Y浓度 ≥ 4% 表示胎儿DNA足够，否则为“浓度不足”或女胎
# 若题目要求“≤ 4%”，请保留；否则建议改为 ≥
# 这里按您原逻辑：保留 Y染色体浓度 ≥ 0.04 的样本（即 >4%）
condition1 = df['Y染色体浓度'] >= 0.04

# 清洗条件2: 孕周在10-25周之间
condition2 = (df['检测孕周_数值'] >= 10) & (df['检测孕周_数值'] <= 25)

# 合并条件
df_cleaned = df[condition1 & condition2].copy()

print(f"满足 Y染色体浓度 ≥ 4% 的样本数: {len(df[condition1])}")
print(f"孕周在10-25周的样本数: {len(df[condition2])}")
print(f"同时满足两个条件的样本数: {len(df_cleaned)}")
```

```python
print("\n清洗后数据缺失情况:")
print(df_cleaned.isnull().sum())

# 对关键字段进行缺失值处理
# 示例：用中位数填充数值型变量，众数填充分类变量
numeric_fill = ['13号染色体的Z值', '18号染色体的Z值', '21号染色体的Z值',
                'X染色体的Z值', 'Y染色体的Z值', '孕妇BMI', '年龄']
category_fill = ['胎儿是否健康', '检测孕周']

for col in numeric_fill:
    if col in df_cleaned.columns:
        median_val = df_cleaned[col].median()
        df_cleaned[col].fillna(median_val, inplace=True)
        print(f"{col} 缺失值已用中位数 {median_val:.2f} 填充")

for col in category_fill:
    if col in df_cleaned.columns:
        mode_val = df_cleaned[col].mode()[0] if not df_cleaned[col].mode().empty else '未知'
        df_cleaned[col].fillna(mode_val, inplace=True)
        print(f"{col} 缺失值已用众数 '{mode_val}' 填充")
```

```python
duplicate_count = df_cleaned.duplicated().sum()
print(f"重复样本数: {duplicate_count}")
if duplicate_count > 0:
    df_cleaned = df_cleaned.drop_duplicates()

df_cleaned.reset_index(drop=True, inplace=True)
print("数据清洗完成，索引已重置。")
```
